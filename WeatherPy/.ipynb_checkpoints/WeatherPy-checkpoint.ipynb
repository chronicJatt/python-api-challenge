```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import requests
import json

# Bring in citipy
from citipy import citipy

# OpenWeather API key
from config import api_key

# OpenWeather API URL & Query URL
qurl = f"https://api.openweathermap.org/data/2.5/weather?appid={api_key}&units=metric&q="

# Lat & Lng Extent
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# Lists
lat_lngs = []
cities = []

# Random generation of coordinates
lat = np.random.uniform(low = -90.000, high = 90.000, size = 1500)
lng = np.random.uniform(low = -180.000, high = 180.000, size = 1500)
lat_lngs = zip(lat, lng)

#Generate city list
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)

# Ensure over 500 cities
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Pre-dataframe lists to house API data
city_name = []
latitude = []
longitude = []
max_temp = []
hum = []
cloudiness = []
wind_sp = []
country = []
date = []

# City count
city_count = 0

# Looping through API calls
for city in cities:
    try:
        response = requests.get(qurl + city.replace(" ","&")).json()
        latitude.append(response['coord']['lat'])
        longitude.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        hum.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_sp.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        city_name.append(city)
        city_count += 1
        print(f"Checking weather for {city} | City No. {city_count}")
    except Exception:
        print(f"{city} not found. Next -->")
print("-----------------------------------------------------")
print(f"Finished API Calls! Total cities found: {city_count}")
print("-----------------------------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Importing lists into dictionary for DataFrame
weather_dict = {
    "City": city_name,
    "Latitude": latitude,
    "Longitude": longitude,
    "Max Temperature (°C)": max_temp,
    "Humidity": hum,
    "Cloudiness": cloudiness,
    "Wind Speed": wind_sp,
    "Country": country,
    "Date": date
}

weather_df = pd.DataFrame(weather_dict)
```

```python
# Output DataFrame to csv
output_data_file = "Output/weather_df.csv"
weather_df.to_csv(output_data_file)
```

## Linear Regression

```python
# Splitting data into hemispheres
north_df = weather_df.loc[pd.to_numeric(weather_df['Latitude']).astype(float) > 0, :]
south_df = weather_df.loc[pd.to_numeric(weather_df['Latitude']).astype(float) < 0, :]

# Compare datasets
print(f'Cities in Northern Hemisphere: {len(north_df)}')
print(f'Cities in Southern Hemisphere: {len(south_df)}')
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Variables
x_var = north_df['Latitude']
y_var = north_df['Max Temperature (°C)']

# Linear regression equation
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_var, y_var)
regress_values = x_var * slope + intercept

# Line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

# Plotting data
plt.scatter(x_var, y_var, edgecolors='black')
plt.plot(x_var, regress_values, 'r-')
plt.annotate(line_eq, (1,-36), fontsize=14, color='red')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (°C)')
plt.title('Northern Hemisphere - Max Temp (°C) vs. Latitude Linear Regression')
plt.savefig('Output/Fig 5 - Northern Hemisphere - Max Temp vs. Latitude Linear Regression.png')
plt.show()

print(f'The r-value is: {rvalue}')
```
